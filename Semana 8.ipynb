```python
from sklearn import tree
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X,y = iris.data, iris.target

tamanho_sepal = []
tamanho_petala = []
fig, ax = plt.subplots()
c = []

for i in iris.target:
    if(i== 0):
        c.append('#cc00ff')
    elif(i== 1):
        c.append('#3333ff')
    elif(i== 2):
        c.append('#ff9966')

for i in X:
    tamanho_sepal.append(i[0])   
    tamanho_petala.append(i[2])

ax.scatter(tamanho_sepal,tamanho_petala,c=c)
plt.show()
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
tree.plot_tree(clf)
```
