```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
label = iris.target_names
X,y = iris.data, iris.target

# A fim de entender melhor o comportamento do código, optei por não aleatoriezar os dados

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

a = clf.predict(X_test)

mtr = confusion_matrix(y_test,a)

disp = ConfusionMatrixDisplay(confusion_matrix= mtr)
disp.plot()
```

```python
print(classification_report(y_test, a, target_names=label))
```
